```python
from overturemaps import core
from lonboard import viz
import pandas as pd
```

```python
# London area centred around Kings Cross
bbox = (-0.134075,51.516082,-0.091708,51.533361)
```

```python
def load_area(overture_type, bbox):
    gdf = core.geodataframe(overture_type, bbox=bbox)
    gdf = gdf.set_crs("EPSG:4326")
    gdf = gdf.clip(bbox)
    return gdf
```

```python
building_gdf = load_area("building", bbox)
```
