```python
from overturemaps import core
from lonboard import viz
```

```python
# London area centred around Kings Cross
bbox = (-0.134075,51.516082,-0.091708,51.533361)
```

```python
def load_area(bbox):
    gdf = core.geodataframe("segment", bbox=bbox)
    gdf = gdf.set_crs("EPSG:4326")
    return gdf
```

```python
gdf = load_area(bbox)
```

```python
gdf.head()
```

```python
gdf.plot(facecolor="#628290", edgecolor="#006064", lw=0.05)
```

```python
viz(gdf)
```

```python
gdf_clipped = gdf.clip(bbox)
```

```python
viz(gdf_clipped)
```
