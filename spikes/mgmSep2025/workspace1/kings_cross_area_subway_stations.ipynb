```python
from overturemaps import core
from lonboard import viz
import pandas as pd
```

```python
# London area centred around Kings Cross
bbox = (-0.134075,51.516082,-0.091708,51.533361)
```

```python
def load_area(overture_type, bbox):
    gdf = core.geodataframe(overture_type, bbox=bbox)
    gdf = gdf.set_crs("EPSG:4326")
    gdf = gdf.clip(bbox)
    return gdf
```

```python
segment_gdf = load_area("segment", bbox)
```

```python
segment_gdf.head()
```

```python
viz(segment_gdf)
```

```python
def find_subway_segment_connectors(gdf):
    subway_segment_gdf = gdf[gdf["class"] == 'subway']
    exploded_gdf = subway_segment_gdf.explode("connectors")
    connectors_df = pd.DataFrame(exploded_gdf["connectors"].tolist(), index=exploded_gdf.index)
    joined_gdf = exploded_gdf.join(connectors_df)
    return joined_gdf
```

```python
subway_segment_gdf = find_subway_segment_connectors(segment_gdf)
subway_segment_gdf.head()
```

```python
viz(subway_segment_gdf)
```

```python
connector_gdf = load_area("connector", bbox)
```

```python
connector_gdf.head()
```

```python
viz(connector_gdf)
```

```python
def find_connector_subset(connector_gdf, subway_segment_gdf):
    connector_subsetted_gdf = connector_gdf.merge(subway_segment_gdf.loc[:,"connector_id"],
        left_on="id",
        right_on="connector_id",
        how="inner"
    )
    connector_subsetted_unique_gdf = connector_subsetted_gdf.drop_duplicates(subset='id')
    return connector_subsetted_unique_gdf
```

```python
connector_subsetted_gdf = find_connector_subset(connector_gdf, subway_segment_gdf)
connector_subsetted_gdf.head()
```

```python
viz(connector_subsetted_gdf)
```
