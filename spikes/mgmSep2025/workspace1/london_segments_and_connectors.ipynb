```python
from overturemaps import core
from lonboard import viz
import pandas as pd
```

```python
# London area centred around Kings Cross
bbox = (-0.134075,51.516082,-0.091708,51.533361)
```

```python
def load_area(overture_type, bbox):
    gdf = core.geodataframe(overture_type, bbox=bbox)
    gdf = gdf.set_crs("EPSG:4326")
    gdf = gdf.clip(bbox)
    return gdf
```

```python
segment_gdf = load_area("segment", bbox)
```

```python
segment_gdf.head()
```

```python
segment_gdf.info()
```

```python
segment_gdf.loc[:,["id","connectors"]]
```

```python
segment_gdf.at[segment_gdf.index[0], "connectors"]
```

```python
segment_gdf_exploded = segment_gdf.explode("connectors")
segment_gdf_exploded.head()
```

```python
connectors_df = pd.DataFrame(segment_gdf_exploded["connectors"].tolist(), index=segment_gdf_exploded.index)
connectors_df.head()
```

```python
segment_gdf_joined = segment_gdf_exploded.join(connectors_df)
segment_gdf_joined.head()
```

```python
viz(segment_gdf)
```

```python
viz(segment_gdf[segment_gdf["class"] == 'subway'])
```

```python
subway_segment_gdf = segment_gdf_joined[segment_gdf_joined["class"] == 'subway']
subway_segment_gdf.head()
```

```python
viz(subway_segment_gdf)
```

```python
connector_gdf = load_area("connector", bbox)
```

```python
connector_gdf.head()
```

```python
viz(connector_gdf)
```

```python
connector_gdf.loc[0, "sources"]
```

```python
connector_subsetted_gdf = connector_gdf.merge(subway_segment_gdf.loc[:,"connector_id"],
    left_on="id",
    right_on="connector_id",
    how="inner"
)
connector_subsetted_gdf.head()
```

```python
viz(connector_subsetted_gdf)
```

```python
connector_subsetted_unique_gdf = connector_subsetted_gdf.drop_duplicates(subset='id')
connector_subsetted_unique_gdf.head()
```

```python
viz(connector_subsetted_unique_gdf)
```
